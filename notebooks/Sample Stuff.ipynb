## Simple Linear Regression
Using the diabetes data set from scikit-learn

```python
## Load the data
from sklearn import datasets
diabetes = datasets.load_diabetes()
```

```python
## Will only use the second feature
diabetes_X = diabetes['data'][:,np.newaxis,2]

## Split the data and targets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]
```

```python
## Create a linear regression model and train it
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
regr.fit(diabetes_X_train,diabetes_y_train)
```

```python
## Make predictions using the test data set
diabetes_y_pred = regr.predict(diabetes_X_test)

## Calculate metric
from sklearn.metrics import mean_squared_error, r2_score
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(diabetes_y_test, diabetes_y_pred))
```

```python
# Plot outputs
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(diabetes_X_test,diabetes_y_test,color='black',s=50,label='Test data')
plt.scatter(diabetes_X_train,diabetes_y_train,color='red',s=15, alpha = 0.3, label='Train data')
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', linewidth=3, label='Trained model')
plt.xlabel('Body mass index (au)')
plt.ylabel('Desease projection (au)')
plt.legend()
plt.show()
```
